#**KMeans Cluster**

Unsupervised Machine Learning is the process of teaching a computer to use unlabeled, unclassified data and enabling the algorithm to operate on that data without supervision. Without any previous data training, the machineâ€™s job in this case is to organize unsorted data according to parallels, patterns, and variations.

K means clustering, assigns data points to one of the K clusters depending on their distance from the center of the clusters. It starts by randomly assigning the clusters centroid in the space. Then each data point assign to one of the cluster based on its distance from centroid of the cluster. After assigning each point to one of the cluster, new cluster centroids are assigned. This process runs iteratively until it finds good cluster. In the analysis we assume that number of cluster is given in advanced and we have to put points in one of the group.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#create a dummy cluster dataset
from sklearn.datasets import make_blobs

X, y = make_blobs(
    n_samples = 300,
    centers = 4,
    cluster_std = 0.60,
    random_state = 0
)
X[:5]
```

```python
#plot a scatter plot
plt.scatter(X[:,0], X[:,1])
```

```python
#train model
from sklearn.cluster import KMeans
model_kmeans = KMeans(4)
```

```python
y_pred = model_kmeans.fit_predict(X)
y_pred[:5]
```

```python
plt.scatter(X[:,0], X[:,1],c=y_pred)
```

```python
#second Example
#create a dummy cluster dataset
from sklearn.datasets import make_blobs

X, _ = make_blobs(
    n_samples = 300,
    centers = 3,
    cluster_std = 0.60,
    random_state = 0
)
X[:5]
```

```python
#split data
from sklearn.model_selection import train_test_split
X_train, X_test, _, _ = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.cluster import KMeans
model_kmeans = KMeans(3)
```

```python
model_kmeans.fit(X_train)
print('Model Trained')
```

```python
y_pred = model_kmeans.predict(X_test)
```

```python
plt.scatter(X_test[:,0], X_test[:,1],c=y_pred)
```

```python
#find cluster id for 1 record
result = model_kmeans.predict( [[-1.57846247,  3.03445802]])
print(result)
```

#**Hierarchical Clustering**

Hierarchical clustering is a connectivity-based clustering model that groups the data points together that are close to each other based on the measure of similarity or distance. The assumption is that data points that are close to each other are more similar or related than data points that are farther apart.

A dendrogram, a tree-like figure produced by hierarchical clustering, depicts the hierarchical relationships between groups. Individual data points are located at the bottom of the dendrogram, while the largest clusters, which include all the data points, are located at the top. In order to generate different numbers of clusters, the dendrogram can be sliced at various heights.

Types of Hierarchical Clustering
Basically, there are two types of hierarchical Clustering:
 - Agglomerative Clustering
 - Divisive clustering

#**Hierarchical Agglomerative Clustering**

It is also known as the bottom-up approach or hierarchical agglomerative clustering (HAC). A structure that is more informative than the unstructured set of clusters returned by flat clustering. This clustering algorithm does not require us to prespecify the number of clusters. Bottom-up algorithms treat each data as a singleton cluster at the outset and then successively agglomerate pairs of clusters until all clusters have been merged into a single cluster that contains all data.

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
#create a dummy cluster dataset
from sklearn.datasets import make_blobs

X, y = make_blobs(
    n_samples = 300,
    centers = 4,
    cluster_std = 0.60,
    random_state = 0
)
X[:5]
```

```python
#plot a scatter plot
plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=4)
```

```python
#Training and prediction
y_pred = ac.fit_predict(X)
print('Predictions Done')
```

```python
plt.scatter(X[:,0], X[:,1],c=y_pred)
```

```python
#plot dendrogram
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize = (8,8))
plt.title('Data')
Dendrogram = sch.dendrogram((sch.linkage(X, method='ward')))
```

#**Credit Card Dataset**

```python
import numpy as np
import pandas as pd
```

```python
cc_ds = pd.read_csv('/content/ml_dataset/CC GENERAL.csv')
cc_ds.head()
```

```python
cc_ds.info()
```

```python
#drop cust_id
cc_ds = cc_ds.drop('CUST_ID', axis=1)
cc_ds.info()
```

```python
#Find sum of all null values
cc_ds.isna().sum()
```

```python
#fill null values with forward record
cc_ds.fillna(method='ffill', inplace=True)
cc_ds.head()
```

```python
cc_ds.info()
```

**Scaling**

to be done so that records won't be biased based on the value size

```python
#scaling the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
cc_ds_scaled = scaler.fit_transform(cc_ds)
cc_ds_scaled[:5]
```

**Normalize**

```python
from sklearn.preprocessing import normalize
cc_ds_normalized = normalize(cc_ds_scaled)
cc_ds_normalized[:5]
```

**Reducing**

```python
#Reduce 18 cols to 2

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
cc_ds_principal = pca.fit_transform(cc_ds_normalized)
cc_ds_principal = pd.DataFrame(cc_ds_principal)
cc_ds_principal.columns=['P1','P2']
```

```python
cc_ds_principal.head()
```

```python
cc_ds_principal.info()
```

```python
#Apply Kmeans
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2)

```

```python
y_pred = km.fit_predict(cc_ds_principal)
```

```python
#plot
plt.scatter(cc_ds_principal['P1'],cc_ds_principal['P2'], c=y_pred)
```

```python
#AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
ac2 = AgglomerativeClustering(n_clusters=2)
```

```python
y_pred = ac2.fit_predict(cc_ds_principal)
```

```python
#plot
plt.scatter(cc_ds_principal['P1'],cc_ds_principal['P2'], c=y_pred)
```
